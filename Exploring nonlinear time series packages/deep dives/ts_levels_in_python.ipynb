```python
# Import the necessary modules and libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
```

```python
# create flatten function to use when you have lists within lists
flatten = lambda l: [item for sublist in l for item in sublist]
```

### Create a python version of the ts_levels function

```python
def ts_levels_py(col, criterion="mse", max_depth=2, min_samples_leaf=1, min_samples_split=2, max_leaf_nodes=30, plot=True):
    # create flatten function
    flatten = lambda l: [item for sublist in l for item in sublist]
    ts = col.to_numpy()
    x = list(np.arange(len(ts)))
    x = np.array(x)
    x = x.reshape(-1,1)
    dtr = DecisionTreeRegressor(criterion=criterion, max_depth=max_depth, min_samples_leaf=min_samples_leaf, 
                                min_samples_split=min_samples_split, max_leaf_nodes=max_leaf_nodes)
    tree = dtr.fit(x,ts)
    p = tree.predict(x)
    if plot == True:
        plt.figure(figsize=(9,7))
        plt.scatter(x, ts, s=20, edgecolor="orange",
            c="cornflowerblue")
        p1 = plt.plot(ts)
        p2 = plt.plot(p)
        plt.ylabel('Amount')
        plt.xlabel('--Time--')
        plt.legend((p1[0], p2[0]), ('Original Time Series', 'Time Series Levels'))
        plt.title("Time Series Levels Plot")
    dfs = pd.DataFrame({"t_steps":flatten(x.tolist()), "original_ts":ts.tolist(), "ts_levels":p.tolist()})
    return dfs
```

### Create a random dataset to play with

```python
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

d = {"X":flatten(X.tolist()), "y":y.tolist()}
df = pd.DataFrame(data=d)
#_ = df["y"].plot()
```

### Run the Ts_levels function

```python
df_ts = ts_levels_py(df["y"], max_depth=2, min_samples_split=20, min_samples_leaf=1, plot=True)
```
